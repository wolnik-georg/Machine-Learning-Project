```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path

def make_grid(
    classes,
    label_map,
    in_dir,
    outfile,
    ext="png",
):
    """
    Create a len(classes) x 3 grid:

        columns: [Correct, Incorrect (true label), Incorrect (pred class)]
        rows:    given class ids in `classes`

    label_map: dict {class_id: "name"} for row labels.
    """
    in_dir = Path(in_dir)

    n_rows = len(classes)
    n_cols = 3
    col_titles = ["Correct", "Incorrect (true label)", "Incorrect (pred class)"]

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(7, 2.0 * n_rows))
    if n_rows == 1:
        axes = axes.reshape(1, -1)

    for row, cls in enumerate(classes):
        correct = sorted(in_dir.glob(f"correct_label{cls}_{cls}-1.{ext}"))
        incorrect_label = sorted(in_dir.glob(f"incorrect_label{cls}_*-1.{ext}"))
        incorrect_pred = sorted(in_dir.glob(f"incorrect_pred*_{cls}-1.{ext}"))

        # take first match of each type (assuming you have at least one)
        imgs = [correct[0], incorrect_label[0], incorrect_pred[0]]

        for col, img_path in enumerate(imgs):
            ax = axes[row, col]
            img = mpimg.imread(img_path)
            ax.imshow(img)
            ax.axis("off")

            if row == 0:
                ax.set_title(col_titles[col], fontsize=9)

        # text label left to the row (class name)
        row_label = label_map.get(cls, str(cls))
        fig.text(
            0.02,
            (n_rows - row - 0.5) / n_rows,
            row_label,
            va="center",
            ha="left",
            fontsize=9,
        )

    # reduce whitespace between images
    plt.subplots_adjust(left=0.12, right=0.99, top=0.93, bottom=0.05,
                        wspace=0.05, hspace=0.05)

    plt.savefig(outfile, dpi=300, bbox_inches="tight")
    plt.close()
    print(f"Saved grid to {outfile}")
```

```python
label_map = {
    9:  "bottle",
    11: "boy",
    39: "keyboard",
    55: "otter",
    68: "road",
    74: "shrew",
}

in_dir = r"C:\Users\YOURNAME\Downloads\selected images\png"

# Grid 1: 55, 11, 74
make_grid(
    classes=[55, 11, 74],
    label_map=label_map,
    in_dir=in_dir,
    outfile="grid_otter_boy_shrew.pdf",
)

# Grid 2: 9, 39, 68
make_grid(
    classes=[9, 39, 68],
    label_map=label_map,
    in_dir=in_dir,
    outfile="grid_bottle_keyboard_road.pdf",
)
```
